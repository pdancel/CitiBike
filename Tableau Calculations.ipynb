```python
import pandas as pd
import numpy as np
# from IPython.core.interactiveshell import InteractiveShell
# InteractiveShell.ast_node_interactivity = "all"
```

```python
all2017_df = pd.read_csv('all2017_df.csv')
all2017_df = all2017_df.drop('Unnamed: 0', axis=1)
all2017_df['Start Hour'] = all2017_df['Start Hour'].fillna(0)
all2017_df.head()
```

## Starting Stations

```python
starting_stations_df = all2017_df[['start station name','start station latitude','start station longitude']]
starting_station_count = starting_stations_df.groupby('start station name')['start station name'].count()
starting_station_count_df = pd.DataFrame(starting_station_count)
starting_station_count_df


starting_stations_df = starting_stations_df.pivot_table(index=['start station name'], values= ['start station latitude','start station longitude'])
starting_stations_df 
```

```python
starting_station_join = starting_station_count_df.join(starting_stations_df,how='inner')
starting_station_join.columns = ['Number of Riders','start station latitude','start station longitude']
starting_station_join.sort_values('Number of Riders', ascending=False)
starting_station_join.to_csv('starting_stations_w_latlng.csv')
```

```python
starting_station_join.head()
```

## Ending Stations

```python
ending_stations_df = all2017_df[['end station name','end station latitude','end station longitude']]
ending_station_count = ending_stations_df.groupby('end station name')['end station name'].count()
ending_station_count_df = pd.DataFrame(ending_station_count)
ending_station_count_df


ending_stations_df = ending_stations_df.pivot_table(index=['end station name'], values= ['end station latitude','end station longitude'])
ending_stations_df 
```

```python
ending_station_join = ending_station_count_df.join(ending_stations_df,how='inner')
ending_station_join.columns = ['Number of Riders','end station latitude','end station longitude']
ending_station_join.sort_values('Number of Riders', ascending=False)
ending_station_join
ending_station_join.to_csv('ending_stations_w_latlng.csv')
```
